1. Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed  in a comma-separated sequence on a single line.

```python
#the lowest of the specified range assigned to n
n = 2000                                      
#list to store all the values which are required
lst = []                                      

#check whether n is lower than or equal to the highest of the specified range
while (n<=3200): 
    #check if n is divisible by 7
    if (n%7==0):
        #check if n is multiple of 5
        if (n%5!=0):
            #adding the required values in the list
            lst.append(n)
    #incrementing n to terminate the while loop
    n+=1                                      

#printing the required values and separting the last term to be ended with "." and other terms with ","
for i in range(len(lst)):
    if (i!=len(lst)-1):
        print(lst[i],end=", ")
    else:
        print(lst[i],end=".")
```

2. Write a Python program to accept the user's first and last name and then getting them printed in the the reverse order with a space between first name and last name.

```python
#using slicing method

#taking inputs
first_name = input()
last_name = input()

#printing the required output
print(first_name[::-1] + " " + last_name[::-1])
```

```python
#using loops

#defined the function reverse to reverse the string
def reverse(s):
    str = "" 
    for i in s: 
        str = i + str
    return str
  
#taking inputs
first_name = input()
last_name = input()

#printing the required output
print (reverse(first_name) + " " + reverse(last_name))
```
